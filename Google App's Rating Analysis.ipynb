### Data Imputation & Manipulation

#### Filling the null values with appropriate values using aggregate functions such as mean,median or mode

##### For categorical value - using mode , For numberical value - will be using the median  and not the mean because data is right skeweed 

```python
#Defining impute_median
def impute_median(series):
    return series.fillna(series.median()) #fillna used to make up for missing values. median is a pandas function
```

```python
google_data.Rating = google_data['Rating'].transform(impute_median)
```

```python
#count the null values again
google_data.isnull().sum()
```

##### Filling up Categorical values

```python
print (google_data['Type'].mode())
print (google_data['Current Ver'].mode())
print (google_data['Android Ver'].mode())
```

```python
google_data['Type'].fillna(str(google_data['Type'].mode().values [0]), inplace = True) # value [0] added to be sure for bimodal scenarios
google_data['Current Ver'].fillna(str(google_data['Current Ver'].mode().values [0]), inplace = True)
google_data['Android Ver'].fillna(str(google_data['Android Ver'].mode().values [0]), inplace = True)
```

```python
#Check for null values again 
google_data.isnull().sum()
```

### Converting Price, Review and Installs to Numerical Values

```python
google_data['Price'] = google_data['Price'].apply(lambda x : str(x).replace('$','') if '$' in str(x) else str(x))
google_data['Price'] = google_data['Price'].apply(lambda x: float(x))
```

```python
google_data['Reviews'] = pd.to_numeric (google_data['Reviews'], errors = 'coerce')
```

```python
google_data['Installs'] = google_data['Installs'].apply(lambda x : str(x).replace('+','') if '+' in str(x) else str(x))
google_data['Installs'] = google_data['Installs'].apply(lambda x : str(x).replace(',','') if ',' in str(x) else str(x))
google_data['Installs'] = google_data['Installs'].apply(lambda x : float(x))
```

## Data Visualization

```python
grp = google_data.groupby('Category')
x = grp['Rating'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)
print (x)
print (y)
print (z)
```

```python
plt.figure(figsize = (16,5))
plt.plot (x, 'ro', color = 'blue')
plt.xticks(rotation= 90)
plt.title('Ratings by Category')
plt.xlabel('Categories')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize = (16,5))
plt.plot (y, 'r--', color = 'blue')
plt.xticks(rotation= 90)
plt.title('Price by Category')
plt.xlabel('Categories')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize = (16,5))
plt.plot (z, 'bs', color = 'blue')
plt.xticks(rotation= 90)
plt.title('Reviews by Category')
plt.xlabel('Categories')
plt.ylabel('Reviews')
plt.show()
```
